```python
from rpolygonpoint.utils.spark import spark
from rpolygonpoint.containerpolygon import ContainerPolygon
from rpolygonpoint.utils.plot import plot_polygon, add_polygon, polygon_to_list
```

```python
# Path to data and workspace
path_ws = "C:/D/engines/rpolygonpoint/rpolygonpoint/data/"
```

```python
# Load data to Mexico polygon
df_mexico_polygon = spark.read.parquet(path_ws + "t_country_polygon")\
    .filter("country_id='MX'")\
    .withColumnRenamed("longitude_id", "coord_x")\
    .withColumnRenamed("latitude_id", "coord_y")

df_mexico_polygon.show(5)
```

```python
# Spark DataFrame to list
lst_mexico_polygon = polygon_to_list(df_mexico_polygon)
```

```python
# Generated Mexico polygon mesh
mexico_polygon_mesh = ContainerPolygon(df_mexico_polygon)

mexico_polygon_mesh.set_mesh_split(2) # Split mesh cells
mexico_polygon_mesh.set_mesh_level(10) # Iterations number to split mesh cells
mexico_polygon_mesh.set_earned_prop(0.1) # earned rate to continue split
mexico_polygon_mesh.set_path_data(path_ws + "mx_territoy_polygonmesh/")

mexico_polygon_mesh.get_polygon_mesh()
# mexico_polygon_mesh.load_polygon_mesh() # Load preprocesor to polygon container
```

```python
df_mexico_polygon_mesh = mexico_polygon_mesh.df_polygon_mesh
df_mexico_polygon_mesh.show(5)
```

```python
fig = plot_polygon(lst_mexico_polygon, tick=1, title="Mexico Polygon Mesh(Split=2, Level=10)",  figsize=(100, 100), alpha=0.8, fontsize=5)

lst_levels = df_mexico_polygon_mesh.selectExpr("array_sort(collect_set(cell_level))").first()[0]


for index, level in enumerate(lst_levels):
    
    df_level = df_mexico_polygon_mesh.filter("cell_level = %s" % level)
    
    # Split cell type of Mexico polygon mesh
    lst_mexico_cell_undecided = polygon_to_list(
        df_polygon=df_level.filter("cell_type = 'undecided'"), 
        polygon_id=["polygon_id", "cell_id"]
    )
    
    lst_mexico_cell_inside = polygon_to_list(
        df_polygon=df_level.filter("cell_type = 'inside'"), 
        polygon_id=["polygon_id", "cell_id"]
    )
    
    add_polygon(fig, lst_mexico_cell_undecided, color="gray", alpha=0.8 - 0.05 * index)
    add_polygon(fig, lst_mexico_cell_inside, color="#00396C", alpha=0.8 - 0.05 * index)
```

```python
# Save plot of Mexico polygon mesh
fig[1].savefig(path_ws + "mx_polygon_mesh.png")
```
